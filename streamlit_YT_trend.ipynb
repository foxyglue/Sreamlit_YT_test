```python
import pandas as pd

data = pd.read_csv('trending.csv')
data.head()
```

```python
data.columns
```

```python
data = data[['title', 'channel_name', 'category_id', 'view', 'like', 'dislike', 'comment', 'trending_time']]
```

```python
data.head()
```

```python
#ubah category_id mjd category_name menggunakan file json
import json

with open("category.json", "rb") as f:
    categories = json.load(f)
```

```python
categories
```

```python
categories = pd.json_normalize(categories, record_path="items")
categories_dict = {idx:title for idx, title in zip(categories['id'], categories['snippet.title'])}
```

```python
data['category'] = [categories_dict[str(x)] if str(x) in list(categories_dict.keys()) else "Unknown" for x in data["category_id"]]
```

```python
data
```

```python
#lihat semua kategori dan jumlahnya
data['category'].value_counts()
```

```python
#cek ada NA atau tidak
data.isnull().sum()
```

```python
#merubah nilai NA mjd 0
data = data.fillna(0)
data.isnull().sum()
```

```python
#merubah trending_time menjadi trending_date

data['trending_time'] = pd.to_datetime(data['trending_time'], format='ISO8601')

# Creating a new column 'trending_date' from 'trending_time' (date part)
data['trending_date'] = data['trending_time'].dt.date

# Dropping the original 'trending_time' column
data = data.drop(['trending_time'], axis=1)
```

```python
data
```

```python
#melihat data duplikat
data[['title', 'channel_name']].value_counts().reset_index(name='counts')
```

```python
#hapus data duplikat
used_data = data.drop_duplicates(['title', 'channel_name'], keep='first')
```

```python
#reset index
used_data = used_data.reset_index(drop= True)
```
